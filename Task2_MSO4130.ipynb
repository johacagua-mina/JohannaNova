```python
df["PECEMPLOYEES"] = (df["ECEMPLOYEES"] / df["SIZE"]) * 100
df.head()
```

## Part (a)

The dataset was successfully imported into Python using pandas.

A new variable, `PECEMPLOYEES`, was created by dividing the number of early-career employees (`ECEMPLOYEES`) by the total company size (`SIZE`) and multiplying by 100.

This represents the percentage of early-career employees in each company.

```python
cont_table = pd.crosstab(df["LONDON"], df["LW"])
cont_table
```

```python
chi2, p, dof, expected = stats.chi2_contingency(cont_table)

print("Chi-square:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:")
print(expected)
```

## Part (b) Interpretation

The chi-square test produced a p-value of 1.0, which is significantly greater than the 0.05 significance level.

Therefore, we fail to reject the null hypothesis.

This indicates that there is no evidence of an association between `LONDON` and `LW` in this dataset.

In other words, whether a company is located in London appears to be independent of whether it pays the Living Wage.

```python
model1 = smf.ols("ABSENT ~ SIZE + TURNOVER + LW", data=df).fit()
print(model1.summary())
```

## Part (c) Regression Interpretation

The regression results show that `LW` (Living Wage) has a statistically significant negative effect on absenteeism.

The coefficient for `LW` is approximately -1.926, with a p-value of 0.000, indicating strong statistical significance.

This suggests that, holding SIZE and TURNOVER constant, firms that pay the Living Wage have, on average, around 1.93 fewer units of absenteeism compared to firms that do not.

In contrast, both SIZE and TURNOVER are not statistically significant, as their p-values are well above 0.05. This indicates that they do not have a meaningful effect on absenteeism in this model.

Overall, the results suggest that paying the Living Wage is associated with lower levels of absenteeism.

The R-squared value of 0.16 suggests that the model explains a relatively small proportion of the variation in absenteeism, indicating that other factors not included in the model may also play an important role.

## Part (f) Correlation and Multicollinearity

The correlation matrix shows that some variables are strongly correlated.

In particular, SIZE and ECEMPLOYEES have a very high positive correlation (approximately 0.93), indicating a strong linear relationship between these variables. TURNOVER is also moderately correlated with both SIZE and ECEMPLOYEES.

High correlation between independent variables can lead to multicollinearity, which is problematic because it makes it difficult to isolate the individual effect of each variable. It can also inflate standard errors and make coefficient estimates less reliable.

Although ECEMPLOYEES is not included as an independent variable in the regression models, there is still moderate correlation between SIZE and TURNOVER (around 0.60), which may introduce some degree of multicollinearity.

To address this issue, we could consider removing one of the correlated variables, combining them into a single measure, or using alternative techniques such as principal component analysis.

Overall, while multicollinearity may be present to some extent, it does not appear severe enough to invalidate the main conclusions of the analysis.

A chi-square test is appropriate in this case because both LONDON and LW are categorical (binary) variables, and we are testing for independence between them.

It is important to note that this analysis is based on observational data, and therefore we cannot infer causality. The results show associations rather than definitive causal relationships.

This may explain why some variables, such as SIZE and TURNOVER, are not statistically significant despite having theoretical relevance.
