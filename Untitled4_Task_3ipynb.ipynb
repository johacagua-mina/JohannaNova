```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import numpy as np

df = pd.read_csv('credit_data_student_M01089122 (2).csv')

# Define independent variables (X) and dependent variable (y)
X = df[['Annual_Income', 'Debt_to_Income', 'Credit_Utilization',
        'Recent_Late_Payments', 'Years_Employed']]
y = df['Default']

# Scale the variables so coefficients are comparable
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit logistic regression model
model = LogisticRegression()
model.fit(X_scaled, y)

# Print coefficients
print("Coefficients (w1 to w5):")
for name, coef in zip(X.columns, model.coef_[0]):
    print(f"  {name}: {coef:.4f}")

print(f"\nIntercept: {model.intercept_[0]:.4f}")

# The threshold: probability > 50% means log-odds > 0
# Credit rating threshold = -intercept
threshold = -model.intercept_[0]
print(f"\nCredit Rating threshold for 50% default probability: {threshold:.4f}")
```

Part (b)

What is logistic regression?
As introduced in the lecture notes on probability distributions, when our outcome variable is binary (here, Default = 0 or 1), we cannot use ordinary linear regression directly. Instead we use logistic regression, which models the probability of default using the logistic function:
$$P(\text{Default}=1) = \frac{1}{1+e^{-\text{CR}}}$$
where CR is the Credit Rating:
$$\text{CR} = w_1 \times \text{Annual_Income} + w_2 \times \text{Debt_to_Income} + w_3 \times \text{Credit_Utilisation} + w_4 \times \text{Recent_Late_Payments} + w_5 \times \text{Years_Employed}$$
The variables were standardised before fitting, using StandardScaler, so that all coefficients are on the same scale and directly comparable to one another.

Interpretation:
All five coefficients have the expected sign and are of a broadly similar magnitude, which makes intuitive sense. Customers with higher debt ratios, higher credit utilisation, and more recent late payments are more likely to default — these all increase the Credit Rating and therefore the predicted probability of default. Conversely, customers with higher annual income and longer employment are less likely to default — these reduce the Credit Rating.
The most influential variable is Recent_Late_Payments (w4 = +0.2601), suggesting that a customer's history of late payments is the strongest single predictor of default. Years_Employed (w5 = −0.2377) is the second most important, reflecting that job stability is a strong indicator of creditworthiness.
The 50% threshold: A customer should be rejected if their Credit Rating exceeds 0.2496. This is because when CR > 0.2496, the logistic function gives P(Default) > 50%, meaning the customer is more likely than not to default.

The key reason for this is that the variables have been standardised before fitting. This means the coefficients reflect not just the raw effect of each variable, but also how much variation there is in that variable across customers. Annual Income has a very large standard deviation (£15,175) compared to the other variables. After standardisation, its effect per standard deviation unit is smaller than that of Recent_Late_Payments or Years_Employed.
Furthermore, in the context of this particular dataset, a customer's payment history and employment stability appear to be more informative about default risk than income alone. A high-income customer who repeatedly makes late payments may still be a poor credit risk. This is consistent with the lecture notes on multivariate distributions, which remind us that variables which appear important individually may have their effect reduced once other correlated variables are included in the model."


Part (d) — Changing the threshold to 20%

"We need to find the Credit Rating threshold at which P(Default) = 20%. Using the logistic function:
$$P(\text{Default}) = \frac{1}{1+e^{-\text{CR}}} = 0.20$$
Solving for CR:
$$1 + e^{-\text{CR}} = 5$$
$$e^{-\text{CR}} = 4$$
$$-\text{CR} = \ln(4)$$
$$\text{CR} = -\ln(4) \approx -1.386$$
However, we must also account for the intercept of −0.2496. The actual threshold in terms of the raw Credit Rating score is:
$$\text{CR}_{\text{threshold}} = -\ln(4) - (-0.2496) = -1.386 + 0.2496 \approx -1.137$$
This means that under the 20% threshold, the bank should reject any customer whose Credit Rating is below −1.137 — a much stricter criterion than the 50% threshold of 0.2496. Many more applications would be rejected under this policy, reducing the bank's exposure to default risk but also turning away potentially good customers.
