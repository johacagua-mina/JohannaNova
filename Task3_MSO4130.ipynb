## Part (c) Interpretation of Annual Income

Although the coefficient for Annual Income appears relatively small, this does not mean that income is unimportant.

One reason is that other variables in the model, such as Debt_to_Income and Credit_Utilization, already capture aspects of a customer's financial situation that are closely related to income.

In particular, the debt-to-income ratio directly incorporates income, meaning that the effect of income is partly absorbed by this variable.

Furthermore, variables such as Recent_Late_Payments reflect actual repayment behaviour, which is often a stronger predictor of default than income alone.

Therefore, while income is relevant, its independent effect appears smaller once we control for more behaviour-based measures of financial risk.

z = ln(p / (1-p))

Therefore, the new rejection threshold for the credit rating is approximately -1.386.

This means that loan applications should be rejected whenever the credit rating is greater than -1.386.

This lower threshold reflects a more risk-averse lending strategy, as the bank is willing to reject customers at a lower predicted probability of default. As a result, more applicants will be rejected compared to the 50% threshold.
