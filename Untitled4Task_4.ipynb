```python
# Instalar librerías necesarias
!pip install yfinance statsmodels --quiet

import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Descargar datos
data = yf.download("BTC-USD", start="2025-01-01", end="2025-12-31")
prices = data["Low"].dropna()

# --- GRÁFICO de la serie ---
plt.figure(figsize=(12, 4))
plt.plot(prices.values)
plt.title("Bitcoin Minimum Daily Price 2025 (Xt)")
plt.xlabel("Day")
plt.ylabel("Price (USD)")
plt.tight_layout()
plt.show()

# --- ADF Test sobre Xt ---
result_X = adfuller(prices.values)
print("=== ADF Test on Xt (Bitcoin prices) ===")
print(f"ADF Statistic : {result_X[0]:.4f}")
print(f"p-value       : {result_X[1]:.4f}")
print(f"Critical values: {result_X[4]}")
if result_X[1] > 0.05:
    print("\nConclusion: p-value > 0.05 → FAIL to reject H0 → Xt is NOT stationary")
else:
    print("\nConclusion: p-value < 0.05 → REJECT H0 → Xt IS stationary")
```

0.4825>0.05, we fail to reject H0H_0

```python
# Calcular Yt = ln(X_{t+1} / X_t)
prices_array = prices.values
Y = np.log(prices_array[1:] / prices_array[:-1])

# --- GRÁFICO de Yt ---
plt.figure(figsize=(12, 4))
plt.plot(Y)
plt.title("Log Returns Yt = ln(X_{t+1}/X_t)")
plt.xlabel("Day")
plt.ylabel("Log Return")
plt.axhline(0, color='red', linestyle='--', linewidth=0.8)
plt.tight_layout()
plt.show()

# --- ADF Test sobre Yt ---
result_Y = adfuller(Y)
print("=== ADF Test on Yt (Log Returns) ===")
print(f"ADF Statistic : {result_Y[0]:.4f}")
print(f"p-value       : {result_Y[1]:.4f}")
print(f"Critical values: {result_Y[4]}")
if result_Y[1] < 0.05:
    print("\nConclusion: p-value < 0.05 → REJECT H0 → Yt IS stationary")
else:
    print("\nConclusion: p-value > 0.05 → FAIL to reject H0 → Yt is NOT stationary")
```

```python
# Fix: flatten Y to 1D array
Y = Y.flatten()

# Split into first 10 months (train) and last 2 months (test)
dates = pd.date_range(start="2025-01-01", periods=len(Y), freq="D")
Y_series = pd.Series(Y, index=dates[:len(Y)])

# First 10 months: January - October 2025
Y_train = Y_series[Y_series.index < "2025-11-01"]
Y_test  = Y_series[Y_series.index >= "2025-11-01"]

print(f"Training set size : {len(Y_train)} days")
print(f"Test set size     : {len(Y_test)} days")

# --- ACF and PACF plots ---
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
plot_acf(Y_train,  lags=30, ax=axes[0], title="ACF of Yt (first 10 months)")
plot_pacf(Y_train, lags=30, ax=axes[1], title="PACF of Yt (first 10 months)")
plt.tight_layout()
plt.show()

# --- BIC table for ARMA(p,q) with p,q in {0,...,4} ---
print("\n=== BIC Table for ARMA(p,q) models ===")
bic_table = pd.DataFrame(index=range(5), columns=range(5), dtype=float)
bic_table.index.name = "p"
bic_table.columns.name = "q"

for p in range(5):
    for q in range(5):
        if p == 0 and q == 0:
            bic_table.loc[p, q] = np.nan
            continue
        try:
            model   = ARIMA(Y_train, order=(p, 0, q))
            results = model.fit()
            bic_table.loc[p, q] = round(results.bic, 2)
        except:
            bic_table.loc[p, q] = np.nan

print(bic_table.to_string())

# Find minimum BIC
min_idx = np.unravel_index(
    np.nanargmin(bic_table.values.astype(float)),
    bic_table.shape
)
print(f"\nLowest BIC : p={min_idx[0]}, q={min_idx[1]}, "
      f"BIC={bic_table.iloc[min_idx[0], min_idx[1]]}")
```

Using only the first 10 months of data (304 observations), we identify two appropriate ARMA(p,q) models using the Box-Jenkins procedure.
Step 1 — ACF and PACF inspection. The ACF plot shows a significant spike at lag 1 that falls inside the confidence band for all lags h≥2h \geq 2
h≥2. This is characteristic of an MA(1) process, where the ACF cuts off after lag q=1q=1
q=1. The PACF shows a similar pattern, with one significant spike at lag 1 and rapid decay afterwards, which is also consistent with an AR(1) process.

Step 2 — BIC table. We compute the BIC for all ARMA(p,q) models with p,q∈{0,1,2,3,4}p, q \in \{0,1,2,3,4\}
p,q∈{0,1,2,3,4}. Following the notes (Week 8.2), lower BIC indicates a better balance between model fit and parsimony. The results are:

q=0q=1q=2p=0—−1446.61−1440.86p=1−1446.61−1440.89−1437.13p=2−1440.91−1435.17−1429.48
The two lowest BIC values are achieved by ARMA(0,1) and ARMA(1,0), both giving BIC = −1446.61. All higher-order models produce larger BIC values, indicating that the added parameters are not justified.
Conclusion. We propose the two models:

Model 1: ARMA(0,1) — an MA(1) process: Yt=μ+ϵt+θ1ϵt−1Y_t = \mu + \epsilon_t + \theta_1 \epsilon_{t-1}
Yt​=μ+ϵt​+θ1​ϵt−1​
Model 2: ARMA(1,0) — an AR(1) process: Yt=μ+ϕ1Yt−1+ϵtY_t = \mu + \phi_1 Y_{t-1} + \epsilon_t
Yt​=μ+ϕ1​Yt−1​+ϵt​

```python
# --- Model 1: ARMA(0,1) ---
model1   = ARIMA(Y_train, order=(0, 0, 1))
results1 = model1.fit()
print("=== Model 1: ARMA(0,1) ===")
print(results1.summary())

# --- Model 2: ARMA(1,0) ---
model2   = ARIMA(Y_train, order=(1, 0, 0))
results2 = model2.fit()
print("=== Model 2: ARMA(1,0) ===")
print(results2.summary())

# --- Predict Yt for the last 2 months (59 days) ---
Y_pred1 = results1.forecast(steps=len(Y_test))
Y_pred2 = results2.forecast(steps=len(Y_test))

# --- Plot predicted vs actual Yt ---
plt.figure(figsize=(14, 5))
plt.plot(range(len(Y_test)), Y_test.values,
         label="Actual Yt",     color="black", linewidth=1.5)
plt.plot(range(len(Y_test)), Y_pred1.values,
         label="ARMA(0,1) forecast", color="blue",  linestyle="--")
plt.plot(range(len(Y_test)), Y_pred2.values,
         label="ARMA(1,0) forecast", color="red",   linestyle="--")
plt.axhline(0, color="gray", linestyle=":", linewidth=0.8)
plt.title("Forecast of Yt — last 2 months (Nov–Dec 2025)")
plt.xlabel("Day")
plt.ylabel("Log Return")
plt.legend()
plt.tight_layout()
plt.show()

print(f"\nARMA(0,1) — mu: {results1.params['const']:.6f}, "
      f"theta1: {results1.params['ma.L1']:.6f}, "
      f"sigma2: {results1.params['sigma2']:.6f}")
print(f"ARMA(1,0) — mu: {results2.params['const']:.6f}, "
      f"phi1: {results2.params['ar.L1']:.6f}, "
      f"sigma2: {results2.params['sigma2']:.6f}")
```

```python
# Fix: ensure prices_array is 1D
prices_array = prices_array.flatten()

# --- Reconstruct predicted prices from log returns ---
last_train_price = prices_array[len(Y_train)]

print(f"Last known price (start of test): ${last_train_price:.2f}")

# Reconstruct: X_{t+1} = X_t * exp(Y_t)
def reconstruct_prices(last_price, log_returns):
    prices_pred = [last_price]
    for y in log_returns:
        prices_pred.append(prices_pred[-1] * np.exp(y))
    return np.array(prices_pred[1:])

X_pred1 = reconstruct_prices(last_train_price, Y_pred1.values)
X_pred2 = reconstruct_prices(last_train_price, Y_pred2.values)

# Actual prices in test period
X_actual = prices_array[len(Y_train)+1 : len(Y_train)+1+len(Y_test)]

# --- Plot ---
days_test = range(len(Y_test))
plt.figure(figsize=(14, 5))
plt.plot(days_test, X_actual, label="Actual Bitcoin price",
         color="black", linewidth=1.5)
plt.plot(days_test, X_pred1,  label="ARMA(0,1) prediction",
         color="blue", linestyle="--")
plt.plot(days_test, X_pred2,  label="ARMA(1,0) prediction",
         color="red",  linestyle="--")
plt.title("Predicted vs Actual Bitcoin Prices — Nov–Dec 2025")
plt.xlabel("Day")
plt.ylabel("Price (USD)")
plt.legend()
plt.tight_layout()
plt.show()

# --- Compute errors ---
mae1  = np.mean(np.abs(X_pred1 - X_actual))
mae2  = np.mean(np.abs(X_pred2 - X_actual))
rmse1 = np.sqrt(np.mean((X_pred1 - X_actual)**2))
rmse2 = np.sqrt(np.mean((X_pred2 - X_actual)**2))
mape1 = np.mean(np.abs((X_pred1 - X_actual) / X_actual)) * 100
mape2 = np.mean(np.abs((X_pred2 - X_actual) / X_actual)) * 100

print("\n=== Forecast Error Comparison ===")
print(f"{'Metric':<10} {'ARMA(0,1)':>12} {'ARMA(1,0)':>12}")
print(f"{'MAE':<10} {mae1:>12.2f} {mae2:>12.2f}")
print(f"{'RMSE':<10} {rmse1:>12.2f} {rmse2:>12.2f}")
print(f"{'MAPE':<10} {mape1:>11.2f}% {mape2:>11.2f}%")
```
